```python
Y = Moscow
X = datas
n = np.size(Moscow)
```

```python
X**2
```

```python
a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2) - np.sum(X)**2) #Formula matemática para o coeficiente angular
```

```python
b = np.mean(Y) - a*np.mean(X) #Fórmula matemática para descobrir o coeficiente linear
```

```python
y = a*X+b
```

```python
np.linalg.norm(Moscow-y) #Regressão linear
```
